```python
import pandas as pd
import pickle
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('heart_disease_data.csv')
df.head()
```

```python
df.info()
```

```python
df['target'].value_counts()
```

```python
#1---> DEFECTIVE HEART
#0---> NORMAL HEART
```

```python
#input and output!
x = df.drop('target', axis=1)
y = df['target']
```

```python
#Train Test Spliting
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state = 42)
scaler =  StandardScaler()
x_train = scaler.fit_transform(x_train) #onlt train data is fit as fit means training!
x_test = scaler.transform(x_test)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
model_train_pred = model.predict(x_train)
model_accuracy = accuracy_score(model_train_pred,y_train)
print(model_accuracy*100,'%')
```

```python
model_test_pred = model.predict(x_test)
model_test_accuracy = accuracy_score(model_test_pred,y_test)
print(model_test_accuracy*100,'%')
```

```python
# Save model
with open("heart_disease_model.pkl", "wb") as f:
    pickle.dump(model, f)

# Save scaler
with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)

print("Model and scaler saved successfully!")
```
